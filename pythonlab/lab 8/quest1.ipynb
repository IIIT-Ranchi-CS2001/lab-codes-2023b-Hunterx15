<!-- Q1 Show party wise seat share for following results of the Assembly Elections 2023 in
Two different pie charts on two different plots. Party with highest percentage should be shown as slightly detached ( show the percentage seat share on each wedge )
Two pie charts as subplots on the same figure object
As a bar chart with party name on X axis and seats won on y axis. Show results of both the states on the same bar plot. Give proper legends
Madhya Pradesh
BJP - Win (163) INC - Win (66) BSP – Win ( 0) Others – Win (1)

INC - Win (69) BJP- Win (115) BSP- Win (2) Others-Win (13) -->


```python
import matplotlib.pyplot as plt

mp_seats = {"BJP": 163, "INC": 66, "BSP": 0, "Others": 1}
raj_seats = {"BJP": 115, "INC": 69, "BSP": 2, "Others": 13}

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

mp_total = sum(mp_seats.values())
raj_total = sum(raj_seats.values())
mp_percentages = {party: (seats / mp_total) * 100 for party, seats in mp_seats.items()}
raj_percentages = {party: (seats / raj_total) * 100 for party, seats in raj_seats.items()}

def max_party_percentage(data):
    return max(data, key=data.get)

mp_labels = [f"{party} ({seats})" for party, seats in mp_seats.items()]
mp_explode = [0.1 if party == max_party_percentage(mp_percentages) else 0 for party in mp_seats]
axes[0].pie(mp_percentages.values(), labels=mp_labels, autopct="%.1f%%", explode=mp_explode, startangle=90)
axes[0].set_title("Madhya Pradesh Assembly Results")

raj_labels = [f"{party} ({seats})" for party, seats in raj_seats.items()]
raj_explode = [0.1 if party == max_party_percentage(raj_percentages) else 0 for party in raj_seats]
axes[1].pie(raj_percentages.values(), labels=raj_labels, autopct="%.1f%%", explode=raj_explode, startangle=90)
axes[1].set_title("Rajasthan Assembly Results")

plt.tight_layout()
plt.show()

fig, ax = plt.subplots(figsize=(10, 6))

x = list(mp_seats.keys())
mp_values = list(mp_seats.values())
raj_values = list(raj_seats.values())

bar_width = 0.4
x_indices = range(len(x))

ax.bar(x_indices, mp_values, width=bar_width, label="Madhya Pradesh", color="blue", alpha=0.7)
ax.bar([i + bar_width for i in x_indices], raj_values, width=bar_width, label="Rajasthan", color="orange", alpha=0.7)

ax.set_xticks([i + bar_width / 2 for i in x_indices])
ax.set_xticklabels(x)
ax.set_xlabel("Parties")
ax.set_ylabel("Seats Won")
ax.set_title("Seats Won in Assembly Elections 2023")
ax.legend()

plt.tight_layout()
plt.show()
```
